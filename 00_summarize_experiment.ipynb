```python
import os
import pandas as pd
```

## Extract information from the first folder

```python
path_dir = r'D:\data\T3737_9_2_Final'
list_files = [x for x in os.listdir(path_dir) if 'FINAL' in x]
```

```python
df1 = pd.DataFrame(list_files,columns=['file_name'])

df1['exp_dir'] = path_dir
df1['im_path'] = [os.path.join(x,y) for x,y in zip(df1.exp_dir,df1.file_name)]
df1['slide'] = 'T3737_9_2'
df1['well'] = [x.split('_')[4] for x in list_files]
df1['exp'] = ['_'.join([x,y]) for x,y in zip(df1.slide,df1.well)]
df1['marker'] = [x.split('_')[6].split('-')[0] for x in list_files]
df1['dye'] = [x.split('_')[5] for x in list_files]
df1['im_round'] = [x.split('_')[3][0] for x in list_files]


df1.loc[df1.dye=='DAPI','marker']  = [f'DNA_{x.zfill(2)}' for x in df1.loc[df1.dye=='DAPI','im_round']]
df1
```

```python
density_dict = {'R000':150,
                'R001':200,
                'R002':150,
                'R003':200,
                'R004':150,
                'R005':200,
                'R006':150,
                'R007':200}

df1['density']  = [density_dict[x] for x in df1.well]
```

```python
donor_dict = {'R000':'D1',
                'R001':'D1',
                'R002':'D1',
                'R003':'D1',
                'R004':'D2',
                'R005':'D2',
                'R006':'D2',
                'R007':'D2'}

df1['donor']  = [donor_dict[x] for x in df1.well]
```

```python
treatment_dict = {'R000':'0_plated',
                'R001':'0_plated',
                'R002':'48_stim',
                'R003':'48_stim',
                'R004':'0_plated',
                'R005':'0_plated',
                'R006':'48_stim',
                'R007':'48_stim'}

df1['treatment']  = [treatment_dict[x] for x in df1.well]
```
